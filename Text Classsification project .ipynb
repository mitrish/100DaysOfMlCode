```python
dict = {}
```

```python
def function(filename,ct):
    ct=ct+1 
    with open(filename, 'r',encoding='utf-8', errors='ignore') as content_file:
        content = content_file.read()
        words = []
        words = content.lower().split()
        for word in words:
            if word not in dict:
                dict[word] = 1
            else:
                dict[word] += 1
    return ct
```
